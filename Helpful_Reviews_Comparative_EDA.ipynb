## Helpful_Rating

### Correlation Matrix

```python
corr_matrix = df.corr()
corr_matrix['Helpful_Rating'].sort_values(ascending=False)
```

Review_Length, Sentence_Length, Word_Length and overall should all be considered as they had nothing to do with the creation of the Helpful_Rating column.

```python
corr_matrix = df_10.corr()
corr_matrix['Helpful_Rating'].sort_values(ascending=False)
```

It's surprising that the correlation for overall is nearly 2.5 times as strong in this subset. The other factors are about the same.

### Scatter Matrix

```python
from pandas.plotting import scatter_matrix

attributes = ['Helpful_Rating', 'Review_Length', 'Sentence_Length', 'Word_Length', 'overall']
scatter_matrix(df[attributes], figsize=(12,8))
```

```python
attributes = ['Helpful_Rating', 'Review_Length', 'Sentence_Length', 'Word_Length', 'overall']
scatter_matrix(df_10[attributes], figsize=(12,8))
```

## Conclusion

Although df has more data points, df_10 has stronger correlations between the target column and predictor columns and should perform better. This makes sense because more votes should give a more accurate helpful score.
